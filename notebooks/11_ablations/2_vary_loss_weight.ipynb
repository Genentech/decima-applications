# Fine-tune Borzoi with varying Poisson loss weights

```python
scripts_dir = '/code/decima/scripts'
save_dir="/gstore/data/resbioai/grelu/decima/20240823/"
```

## Parameters

```python
lr = 3e-5
weights = [0.000001, 0.00001, 0.001, 0.01, 0.1]
grad = 5
bs = 4
```

## Generate training commands

```python
gpus = [2,3,4, 5, 6]
for weight, gpu in zip(weights, gpus):
    name=f'decima_v20240823_rep0_weight_{weight}'    
    cmd = f'CUDA_VISIBLE_DEVICES={gpu} python {scripts_dir}/finetune.py --name {name} \
    --dir {save_dir} --lr {lr} --weight {weight} --grad {grad} --replicate 0 \
    --bs {bs}'
    print(cmd)
```
